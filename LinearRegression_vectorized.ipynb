```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def predict(x,w,b):
    return w*x+b
```

```python
def compute_cost(x,y,w,b):
    m=x.shape[1]
    cost=np.sum(((w*x+b)-y)**2,axis=1,keepdims=False)
    cost=cost/(2*m)
    return cost
```

```python
def compute_gradient(x,y,w,b):
    m=x.shape[1]
    dj_dw=np.sum(((w*x+b)-y)*x,axis=1,keepdims=False)
    dj_db=np.sum(((w*x+b)-y),axis=1,keepdims=False)
    dj_dw/=m
    dj_db/=m
    return dj_dw,dj_db
    
```

```python
def gradient_descent(x,y,w_in,b_in,alpha,itera):
    w=w_in
    b=b_in
    j_cost=[]
    j_iter=[]
    for i in range(itera):
        dj_dw,dj_db=compute_gradient(x,y,w,b)
        w=(w-(alpha*dj_dw))
        b=(b-(alpha*dj_db))
        if i%1000==0 or i==itera-1:
            cost_temp=compute_cost(x,y,w,b)
            # print(f"cost: {}");
            j_cost.append(cost_temp)
            j_iter.append(i)
    return w,b,j_cost,j_iter
```

```python
# rather than declaring x_train as 1D array declare it as 2D array so that we can perform vectorization
# which reduces the time for calculations done under for loops:
x_train=np.array([[1.0,2.0,5.0,3.5]])
y_train=np.array([[110.0,220.0,550.0,385.0]])
w_in=0.0
b_in=0.0
alpha=0.07
it=10000
w_f,b_f,j_cost,j_iter=gradient_descent(x_train,y_train,w_in,b_in,alpha,it)
plt.plot(j_iter,j_cost)
print(w_f*x+b_f)
```
